```python
# Alex's unit tests
import unittest

class SpellDamage_Test(unittest.TestCase):
    def setUp(self):
        self.Fireball = spells("Fireball", 30, "fire")
        self.waterblast = spells("Waterblast", 30, "water")

        self.dragon = enemy("Dragon", 100, "fire", 20, 10)

    def test_Fireball(self):
        """Fire should deal extra damage because dragon is weak to fire
        30*1.5 = 45 damage, 100 health - 45 damage should leave dragon with 55 health"""
        self.Fireball.attack(self.dragon)
        self.assertEqual(55,self.dragon.health)

        """waterblast is listed also as 30 damage but the dragon is not weak to water so
        no extra damage is done.
        55 remaining health - 30 damage = 25 health after waterblast"""
        self.waterblast.attack(self.dragon)
        self.assertEqual(25,self.dragon.health)


        
            
```
